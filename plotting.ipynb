```python
import matplotlib.pyplot as plt
import json

# Reading data from JSON files
with open('fpgs.json', 'r') as file:
    data1 = json.load(file)

with open('bim.json', 'r') as file:
    data2 = json.load(file)

# Extracting keys and values
keys = list(data1.keys())  # Assuming both JSON objects have the same keys
values1 = list(data1.values())
values2 = list(data2.values())

# Converting keys to floats for plotting
x = [float(key) for key in keys]

# Plotting
plt.plot(x, values1, 'o-', color='blue', label='FGSM')  # Blue line with circle markers for the first JSON
plt.plot(x, values2, 'o-', color='red', label='Iterative FSGM')   # Red line with circle markers for the second JSON

# Labeling the axes
plt.xlabel('eps')
plt.ylabel('accuracy')

# Adding a legend
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
